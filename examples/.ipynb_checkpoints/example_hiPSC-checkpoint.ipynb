```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import average_precision_score, precision_recall_curve
from sklearn.metrics import auc

from cycler import cycler
plt.style.use('default')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams["figure.figsize"] = (5, 3)

ccycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
#ccycle = cycler(color=['#4E79A7', '#F28E2B', '#E15759', '#76B7B2','#59A14E',
 #                                                '#EDC949','#B07AA2','#FF9DA7','#9C755F','#BAB0AC']).by_key()['color']
    
    
def calc_prec_rec(y, x):
    y = np.abs(y)
    n_reg = (y == 1).sum()*2
    zero_idxs = np.argsort(x)[::-1][n_reg:]
    one_idxs = np.argsort(x)[::-1][:n_reg]
    x[zero_idxs] = 0
    x[one_idxs] = 1
    tp = ((y==1)&(x==1)).sum()
    prec = tp / (x==1).sum()
    rec = tp / (y==1).sum()
    return prec, rec


def calc_auroc_aupr(y, x):
    y = np.abs(y)
    auroc = roc_auc_score(y, x)
    prec, rec, th = precision_recall_curve(y, x)
    #aupr = auc(rec, prec)
    aupr = average_precision_score(y, x)
    return auroc, aupr

def calc_auroc_aupr_ep(y, x):
    y = np.abs(y)
    auroc = roc_auc_score(y, x)
    prec, rec, th = precision_recall_curve(y, x)
    aupr = average_precision_score(y, x)
    
    k = int(y.sum())
    top_k_idx = np.argsort(x)[::-1][:k]
    ep = y[top_k_idx].sum()/k
    
    if -1 in y:
        print('negative y')

    return auroc, aupr, ep
```

```python
ChIP_B = pd.read_csv('./data_201208/cr_gRNA_WT_eiraku/ChIP_B_d2.csv', index_col=0)
chip_genes = list(ChIP_B.dropna(axis=1).columns)

B_renge_0 = pd.read_csv('./data_201208/cr_gRNA_WT_eiraku/B_renge_d2_0.csv', index_col=0)
ko_genes = B_renge_0.index
tf_inc = set(B_renge.index) - set(ko_genes)
tf_chip = list(set(tf_inc) & set(chip_genes))

B = B_renge.copy()
for idx in B_renge.index:
    B.loc[idx][idx] = np.nan
```

```python
cell_class = pd.read_csv('./data_201208/ChIP_Atlas/celltypeList.tab', sep='\t')
cell_class = cell_class.loc[cell_class['Genome'] == 'hg38']
cell_class = cell_class.loc[cell_class['Cell_type_class'] == 'Pluripotent stem cell']

use_cell_types = list(set(cell_class['Cell_type'].values))
use_cell_types = [s.replace(' ', '_') for s in use_cell_types]

use_cell_types = [s for s in use_cell_types if ('derived' not in s)]
```

```python
E = pd.read_csv('./data_201208/cr_gRNA_WT_eiraku/E_renge_d2_80.csv', index_col=0)
target_all_list = list(set(E.columns) - set(['CTRL', 'AAVS1']))
#use_cell_types = ['iPS_cells', 'hESC_HUES64', 'hESC_H9', 'hESC_H1', 'Embryonic_Stem_Cells']


def calc_chip_B(th_chip):
    n_chip = 0

    ChIP_B = pd.DataFrame(np.zeros((len(target_all_list), len(target_all_list))))
    ChIP_B.index = target_all_list
    ChIP_B.columns = target_all_list
    ChIP_B_label = ChIP_B.copy()

    chip_genes = []


    for target in target_all_list:
        try:
            C = pd.read_csv('./data_201208/ChIP_Atlas/'+target+'.10.tsv', sep='\t', index_col=0)
        except Exception as e:     # target gene has no ChIP-seq data
            ChIP_B[target] = np.nan
        else:
            n_chip += 1
            ## cell type specific ChIP
            col = [c.split('|')[1]for c in C.columns[:-1]]
            included_col = []
            for c in col:
                if c in use_cell_types:
                    included_col.append(True)
                else:
                    included_col.append(False)

            included_col.append(False)

            if np.sum(included_col) == 0:   ## no chip data for pluripotent stem cell 
                ChIP_B[target] = np.nan
                continue
        
            included_row = C.index.isin(target_all_list)
            C_pp = C.iloc[included_row, included_col].max(axis=1)
        
            chip_genes.append(target)
    
            for gene in C_pp.index:
                ChIP_B.loc[gene, target] = C_pp.loc[gene]
                if C_pp.loc[gene] > th_chip:
                    ChIP_B_label.loc[gene, target] = 1

    if len(chip_genes) != 20:
        print('len(chip_genes) = '+str(len(chip_genes)))

    return (ChIP_B_label, ChIP_B)
```

```python
if B_renge_pval.min().min() == 0:
    min_p = np.sort(np.array(list(set(B_renge_pval.values.flatten()))))[1]
    B_renge_pval[B_renge_pval == 0] = min_p
```

```python
fig = plt.figure(figsize=(10,8))
fig.subplots_adjust(wspace=0.3, hspace=0.5)

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2,sharey=ax1)

for reg in ['ko', 'non_ko']:
    if reg == 'non_ko':
        gene_list = tf_chip
    else:
        gene_list = list(set(chip_genes) - set(tf_inc))

    ChIP_B_label, ChIP_B = calc_chip_B(th_chip=1)
    max_score = ChIP_B[gene_list].max().max()
    th_chip_list = np.linspace(0, max_score, num=100, endpoint=False)

    
    renge_aupr_list = []
    random_aupr_list = []

    for th_chip in th_chip_list:
        ChIP_B_label, ChIP_B = calc_chip_B(th_chip)
    
        x = B[gene_list].sort_index().values.flatten()
        mask = np.where(~np.isnan(x))[0]
    
        y = ChIP_B_label[gene_list].sort_index().values.flatten()[mask]
        y = np.nan_to_num(y)
    

        x = -np.log10(B_renge_pval[gene_list].sort_index()).values.flatten()[mask]
        #x[B_renge[gene_list].sort_index().values.flatten()[mask] < 0] = 0

        renge_auroc, renge_aupr, renge_ep = calc_auroc_aupr_ep(y, x)
        renge_aupr_list.append(renge_aupr)
        
        random_aupr_list.append((np.abs(y) == 1).sum() / len(y))

    if reg == 'ko':
        renge_aupr_list = np.array(renge_aupr_list) / np.array(random_aupr_list)
 
        ax1.plot(th_chip_list, np.log2(renge_aupr_list), linewidth=3, label='RENGE', color=ccycle[0], zorder=6)
        ax1.hlines(0, th_chip_list[0], th_chip_list[-1], label='random', color='black', linestyle = 'dashed', zorder=7, linewidth=3)


        ax1.set_xlabel('ChIP threshold')
        ax1.set_ylabel('Log2(AUPRC ratio)')
        ax1.set_title('KO gene')
            
    else:
        renge_aupr_list = np.array(renge_aupr_list) / np.array(random_aupr_list)

        ax2.plot(th_chip_list, np.log2(renge_aupr_list), linewidth=3, label='RENGE', color=ccycle[0], zorder=6)
        ax2.hlines(0, th_chip_list[0], th_chip_list[-1], label='random', color='black', linestyle = 'dashed', zorder=7, linewidth=3)


        ax2.set_xlabel('ChIP threshold')
        ax2.set_ylabel('Log2(AUPRC ratio)')
        ax2.set_title('Non-KO gene')
        
        handles, labels = ax1.get_legend_handles_labels()   # show legend labels of ko genes

        #ax2.legend(handles[::-1], labels[::-1], fontsize=10)
        ax2.legend(handles, labels, fontsize=12, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```
